# Importing Libraries

 # Importing the Dataset

```python
tech_skills = pd.read_csv('Example_Technical_Skills.csv')
raw_skills = pd.read_csv('Raw_Skills_Dataset.csv')
```

# Data Preprocessing Accoding to Question

# Renaming the column names in both datasets to common name

```python
raw_skills.rename(columns = {'RAW DATA':'Technology_Skills'}, inplace = True)
```

```python
tech_skills.rename(columns = {'Technology Skills':'Technology_Skills'}, inplace=True)
```

```python
raw_skills
```

```python
tech_skills
```

# Merging both datasets to have common rows

```python
data = raw_skills.merge(tech_skills, how = 'inner' ,indicator=False)
```

```python
data
```

# Droping the duplicates in dataset

```python
data.drop_duplicates(subset ="Technology_Skills", inplace = True)
```
